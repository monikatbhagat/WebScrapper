## WebScrapper

This WebScrapper scrap quotes and other information from www.brainyquote.com webssite

```python
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uReq
```

```python
searchString = input("Enter you ant to search")
```

```python
url = 'https://www.brainyquote.com/search_results?q=' + searchString
```

```python
url
```

```python
response = requests.get(url)
soup = bs(response.text, 'html.parser')
```

```python
quotes = soup.find('div', id='quotesList')
```

```python
print(quotes)
```

```python
allDiv=quotes.find_all('div', {'class': "grid-item qb clearfix bqQt"})
```

```python
quotesList = []
for quote in allDiv:
    Quote=quote.find('a',{'title':"view quote"}).text
    Writer=quote.find('a',{'title':"view author"}).text
    quoteDict={'Quote':Quote,'Writer':Writer}
    quotesList.append(quoteDict)
```
